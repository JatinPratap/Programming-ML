```python
citywise = df.groupby(['City', "Complaint Type"]).size().sort_values(ascending = False).to_frame().reset_index()
citywise.columns = ['City', 'Complaint Type', 'Count of Cases']
```

```python
citywise
```

3.3) types of complaints in each city in a separate dataset

```python
citywise[:20]
```

```python
results = pd.DataFrame(columns=['City', 'Count of Cases', 'Complaint Type'])
```

```python
# Get the unique cities
cities = citywise['City'].unique()

for city in cities:

    # Get the max count of cases for the city
    max_count = citywise[citywise['City'] == city]['Count of Cases'].max()

    # Get the complaint type with the max count of cases
    complaint_type = citywise[citywise['City'] == city]['Complaint Type'].loc[citywise[citywise['City'] == city]['Count of Cases'].idxmax()]
    results = results.append({'City': city, 'Count of Cases': max_count, 'Complaint Type': complaint_type}, ignore_index=True)
```
