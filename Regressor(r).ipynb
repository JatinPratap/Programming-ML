# Cleanup tweets and analyze top hashtags

```python
# import the data
import pandas as pd
```

```python
tweet_data=pd.read_csv('raw_tweets_economy.txt',sep='\t')
```

```python
tweet_data
```

```python
# drop the first col
tweet_data=tweet_data.drop('Unnamed: 0',axis=1)
```

```python
tweet_data.head()
```

```python
tweet_data.rename({'0':'tweet'},axis=1,inplace=True)
```

```python
type(tweet_data['tweet'])
```

```python
tweet_list=tweet_data['tweet'].values
```

# task
1. Normalize the case
2. User Handles @
3. URL s
4. Extract Hashtags

```python
# normalize the case
tweet_lower=[txt.lower() for txt in tweet_list]
```

```python
tweet_lower[:5]
```

```python
import re
```

```python
# remove the @
re.sub('\@','','hello @user1 also @user2')
```

```python
# remove the @
re.sub('\@\w+','','hello @user1 also @user2')
```

```python
tweet_nouser=[re.sub('\@\w+','',twt) for twt in tweet_lower]
```

```python
tweet_nouser[:5]
```

```python
# remove URLs-https ://
re.sub('\w+://\S+','','Rita is learning a course https://t.co/jxet7hqu4k website')
```

```python
# remove URLs-https ://
re.sub('https://\S+','','Rita is learning a course https://t.co/jxet7hqu4k website')
```

```python
tweet_nourl=[re.sub('https://\S+','',twt) for twt in tweet_nouser]
```

```python
tweet_nourl[:5]
```

# extract the hashtags

```python
re.findall('#\w+','I like #games #mobilegames')
```

```python
hastags=[re.findall('#\w+',twt) for twt in tweet_nourl]
```

```python
hastags[:5]
```

```python
len(hastags)
```

```python
# combine the all hashtags
all_hashtags=[]
for twt in hastags:
    all_hashtags.extend(twt)
```
